```python
import re
import requests
from bs4 import BeautifulSoup
import html2text
from datetime import datetime

def getBondReports(startdate='2005-01-01', enddate='2017-12-31', page_cnt=104, savefolder='data/bondreports/pdf/') :
    page_url_format = "https://finance.naver.com/research/debenture_list.nhn?keyword=&brokerCode=&searchType=writeDate&writeFromDate={}&writeToDate={}&x=27&y=20&page={}"
    
    for pagenum in range(1,page_cnt+1) :
        page_url = page_url_format.format(startdate, enddate, pagenum)
        page = requests.get(page_url)
        
        soup = BeautifulSoup(page.content, 'html.parser')
        tr_list = soup.find('table', {"class":'type_1'}).find_all('tr')
        for tr in tr_list :
            td_list = tr.find_all('td')
            if( len(td_list) > 1 ) :
                title = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', td_list[0].text.strip())
                sec = td_list[1].text
                pdf_link = td_list[2].a['href']
                date = td_list[3].text
                filepath = savefolder + "{}_{}_{}.pdf".format(date,sec,title)
                file_res = requests.get(pdf_link)
                print(filepath)
                
                with open(filepath, 'wb') as f:
                    f.write(file_res.content)

getBondReports()    
```

```python
import re
from tika import parser #https://github.com/chrismattmann/tika-python
import shutil
from os import listdir
from os.path import isfile, join

# 폴더내 pdf파일을 txt 파일로 변경
def pdf2txt (source_folder="./data/bondreports/pdf/", output_folder="./data/bondreports/text/") :
    # 지정 폴더 내 파일 목록 조회 (파일만)
    pdf_files = [f for f in listdir(source_folder) if isfile(join(source_folder, f))]
    
    for pdf in pdf_files :
        pdf_filepath = source_folder + pdf
        pdf_tmp_filepath = output_folder + 'tmp.pdf'
        # pdf 파일을 text로 변환
        
        shutil.copyfile(pdf_filepath, pdf_tmp_filepath)
            
        parsedPDF = re.sub('\n', '', parser.from_file(pdf_tmp_filepath)["content"])
        #print(parsedPDF)

        ouput_filepath = (output_folder + pdf).replace('.pdf', '.txt')
        with open(ouput_filepath, 'w') as f :
            print(ouput_filepath)
            f.write(parsedPDF)
            f.close()      
            
pdf2txt()            
```

```python
import csv
import pandas as pd
from os import listdir
from os.path import isfile, join

def txt2csv(source_folder="./data/bondreports/text/", output_file="./data/bondreports/bondreports.csv") :
    # 지정 폴더 내 파일 목록 조회 (파일만)
    txt_files = [f for f in listdir(source_folder) if isfile(join(source_folder, f))]
    df = pd.DataFrame(columns=['date', 'securities', 'title', 'report'])
    
    print(len(txt_files))
    
    for txt_file in txt_files :
        print(txt_file)
        if txt_file[-3:] != 'txt' :
            continue
        
        try :
            with open(source_folder + txt_file, 'r') as f :
                txt = f.read().strip()
                date = txt_file.split('_')[0]
                securities = txt_file.split('_')[1]
                title = txt_file.split('_')[2].split('.')[0]
                df.loc[len(df)] = [date, securities, title, txt]
        except :                
            print("error : " + txt_file)
    
    df.to_csv(output_file, index=False)
        
    
txt2csv()    
```

```python
#http://openuiz.blogspot.com/2016/07/mecab-ko-dic.html
import pandas as pd
from ekonlpy.sentiment import MPCK

def text2ngram(text) :
    mpck = MPCK()
    bef_tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(bef_tokens)
    tokens = []

    #print(tokens)
    print(ngrams)
    #break;

    stoppos = ['SC','SY','SF','SE','SS','SP','SO','SW','SSC','JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JX','JC','EF','EC','ETN','ETM']
    stopword = ['']

    for tag in bef_tokens:
        if tag[1] not in stoppos:
            if tag[0] not in stopword:
                tokens.append(tag)
                
    return tokens, ngrams
```

```python
#http://openuiz.blogspot.com/2016/07/mecab-ko-dic.html
import pandas as pd
from ekonlpy.sentiment import MPCK


def Preprocessing(input_file="./data/bondreports/bondreports.csv", output_file="./data/bondreports/bondreports_preprocessed.csv") :
    df = pd.read_csv(input_file)
    df['tokens'] = ''
    df['ngram'] = ''

    #return 
    for row_idx in df.index: 
        text = df['report'][row_idx]
        
        try : 
            tokens,ngrams = text2ngram(text)

            #print(word)
            #print(",".join(word))
            df['tokens'][row_idx] = ",".join(tokens)
            df['ngram'][row_idx] = ",".join(ngrams)
            print(str(row_idx+1) +'/'+ str(df.shape[0]) +'번째 파일 전처리 완료')
            #print( df['preprocessed'][row_idx] )
        except Exception as e:
            print("============ {} ===========".format(str(e)))
            print("Error :" + text)
        
    df.to_csv(output_file, index=False)
        
Preprocessing()
```
