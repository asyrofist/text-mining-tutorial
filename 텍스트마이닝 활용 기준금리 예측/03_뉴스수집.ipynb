```python
import csv
import pandas as pd
from os import listdir
from os.path import isfile, join

def txt2csv(source_folder="./data/news/text/", output_file_format="./data/news/news_{}.csv") :
    # 지정 폴더 내 파일 목록 조회 (파일만)
    txt_files = [f for f in listdir(source_folder) if isfile(join(source_folder, f))]
    txt_files.sort()
    df = pd.DataFrame(columns=['media','date','title','url', 'news'])
    
    #print(txt_files)
    prev_year = '2005'
    cnt = 1
    for txt_file in txt_files :
        print("{} : {}".format(cnt, txt_file))
        
        year = txt_file[5:9]
        #if year != '2017':
        #    continue
        #print(prev_year)
        
        if prev_year != year :
            df.to_csv(output_file_format.format(prev_year), index=False)
            df = pd.DataFrame(columns=['media','date','title','url', 'news'])
            
        with open(source_folder + txt_file, 'r') as f :
            news_list = f.read().strip().replace("\n","").split("===")
            for news in news_list :
                if len(news.split("@@@")) == 5 : 
                    df.loc[len(df)] = news.split("@@@")
                    
        if txt_file == txt_files[-1:] :
            df.to_csv(output_file_format.format(prev_year), index=False)
            df = pd.DataFrame(columns=['media','date','title','url', 'news'])
                    
        prev_year = year
        
        cnt += 1
        
    
txt2csv()    
```

```python
from datetime import timedelta, date

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2013, 1, 1)
end_date = date(2013, 1, 1)
for single_date in daterange(start_date, end_date):
    print(single_date.strftime("%Y.%m.%d"))
```

```python
import pandas as pd
from ekonlpy.sentiment import MPCK

def text2ngram(text) :
    mpck = MPCK()
    bef_tokens = mpck.tokenize(text)
    ngrams = mpck.ngramize(bef_tokens)
    tokens = []

    #print(tokens)
    #print(ngrams)
    #break;

    stoppos = ['SC','SY','SF','SE','SS','SP','SO','SW','SSC','JKS','JKC','JKG','JKO','JKB','JKV','JKQ','JX','JC','EF','EC','ETN','ETM']
    stopword = ['']

    for tag in bef_tokens:
        if tag[1] not in stoppos:
            if tag[0] not in stopword:
                tokens.append(tag)
                
    return tokens, ngrams
```

```python
import re 

for year in range(2014,2017) : 
    source_folder = "./data/news/"
    df = pd.read_csv(source_folder + 'news_{}.csv'.format(year))
    df['tokens'] = ''
    df['ngram'] = ''

    for i, row in df.iterrows() :
        try :
            text = row['news']
            start_pattern = text[:20]

            #print("========================================")
            #print(text)

            dup_loc = text.rfind(start_pattern)
            if dup_loc > len(start_pattern) :
                text = text[:dup_loc]


            spe_loc = text.find("▶ 이데일리ON")
            if dup_loc > 0 :
                text = text[:spe_loc]

            text = re.sub("(\w+\.)*\w+@(\w+\.)+[A-Za-z]+", '', text)

            author_loc = re.search('(기자 =)', text)
            if author_loc != None :
                text = text[:author_loc.start()].strip()

            copyright_loc = re.search('Copyright', text)
            if copyright_loc != None :
                text = text[:copyright_loc.start()].strip()

            tokens,ngrams = text2ngram(text)

            row['tokens'] = ','.join(tokens)
            row['ngram'] = ','.join(ngrams)
            row['news'] = text
            if i%1000 == 0 :
                print('{} : {}/{}'.format(year, i, df.shape[0]))    
        except Exception as ex:
            print("============== {} : {} ==============".format(i, str(ex)))
            print(row['news'])
            
    save_file = source_folder+ 'news_{}_preprocessed_n.csv'.format(year)
    df.to_csv(save_file, index=False)
```
